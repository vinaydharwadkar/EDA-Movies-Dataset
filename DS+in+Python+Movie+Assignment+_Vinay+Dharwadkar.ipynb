-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.
2. Group the dataframe using `genre_1` as the primary column and `genre_2` as the secondary column.
3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Write your code for extracting the first two genres of each movie here

# Function to extract the first 2 genre
def genres_1_2(genres):
    '''genres_1_2(genres) : function to extract the first two genres
       Input : movies.genres series from movies dataframe
       Return : A series with a list containing two values
              : First values corresponding to 'genre_1' and second one to 'genre_2'
    '''
    # Since the 'genres' series is of type 'object', it is represented as a string.
    # Split the input 'genres' series on "|" symbol using split('|')
    # This results in a list of elements each representing the various genre mentioned in 'genres' series.
    list_genres=genres.split('|')
    
    # Check the length of 'list_genres'. 
    if len(list_genres) > 1:                              # Movie has more than 1 genre identified against it.
        return pd.Series([list_genres[0], list_genres[1]])
    else:                                                 # Movie has only 1 genre which needs to be propogated to 'genre_2'.
        return pd.Series([list_genres[0], list_genres[0]])

movies[['genre_1','genre_2']] = movies['genres'].apply(genres_1_2)
```

```python
# Display head of 'movies' dataframe to confirm inclusion of columns 'genre_1' and 'genre_2'.
movies.head()
```

```python
# Write your code for grouping the dataframe here
# Group the 'movies' dataframe using 'genre_1' and primary and 'genre_2' as secondary grouping column using group_by() function.
movies_by_genre = movies.groupby(['genre_1', 'genre_2'])
```

```python
# Write your code for getting the 5 most popular combo of genres here
# movies_by_genre.gross.mean() : Find mean of 'gross' column grouped by 'genre_1' and 'genre_2'
# sort_values(ascending = False) : sort the resultant series in descending order
# [:5] : Selects only first 5 rows.
PopGenre = movies_by_genre.gross.mean().sort_values(ascending = False)[:5]

#Convert 'PopGenre' series to dataframe , reset index and give meaningful column names
PopGenre = PopGenre.to_frame().reset_index().\
                      rename(columns={'genre_1':'Genre_1','genre_2':'Genre_2','gross':'Avg_Gross_Collection'})

# Display 'PopGenre' dataframe and check which is the most popular genre combo
PopGenre
```

```python
# Write the code for finding the mean of critic reviews and audience reviews here

# Actor by Critic Review
# by_actor_1_name.num_critic_for_reviews.mean() : Find mean of 'num_critic_for_reviews' column grouped by 'actor_1_name'
# sort_values(ascending = False) : sort the resultant series in descending order
actor_critic_review = by_actor_1_name.num_critic_for_reviews.mean().sort_values(ascending = False)

#Convert 'actor_critic_review' series to dataframe , reset index and give meaningful column names
actor_critic_review = actor_critic_review.to_frame().reset_index().\
                      rename(columns={'actor_1_name':'Actor_Name','num_critic_for_reviews':'Avg_Critic_Review'})
#-------------------------------------------------------------------------------------------------------------------#
# Actor by Audience Review
# by_actor_1_name.num_user_for_reviews.mean() : Find mean of 'num_user_for_reviews' column grouped by 'actor_1_name'
# sort_values(ascending = False) : sort the resultant series in descending order
actor_audience_review = by_actor_1_name.num_user_for_reviews.mean().sort_values(ascending = False)

#Convert 'actor_audience_review' series to dataframe , reset index and give meaningful column names
actor_audience_review = actor_audience_review.to_frame().reset_index().\
                        rename(columns={'actor_1_name':'Actor_Name','num_user_for_reviews':'Avg_Audience_Review'})
```
